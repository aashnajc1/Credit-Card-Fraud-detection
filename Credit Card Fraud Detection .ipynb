### The given data is unbalanced since one class(0) has 284315 records whereas the other class(1) has 492 records. To overcome this situation of unbalanced data we perform Under-Sampling of data
We will build a dataset containing similar distributing of both classes
(No of Fraud Transactions -> 492)

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X = new_data.drop(columns = 'Class', axis = 1)
```

```python
Y = new_data['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, stratify = Y, random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
prediction_train = model.predict(X_train)
```

```python
accuracy_train = accuracy_score(prediction_train, Y_train)
```

```python
prediction_train
```

```python
accuracy_train
```

```python
prediction_test = model.predict(X_test)
```

```python
accuracy_test = accuracy_score(prediction_test, Y_test)
```

```python
prediction_test
```

```python
accuracy_test
```

```python
print("Accuracy of train data set : ", round(accuracy_train*100,2))
print("Accuracy of test data set : ", round(accuracy_test*100,2))
```
